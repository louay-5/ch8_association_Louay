```
import pandas as pd               #For data preparation and preprocessing

import matplotlib.pyplot as plt   #For data viz  
import seaborn as sns             #For data viz  

from mlxtend.preprocessing import TransactionEncoder     #For one hot encoder
from mlxtend.frequent_patterns import apriori            #For applying apriori algorithm
from mlxtend.frequent_patterns import association_rules  #For confidence and lift metrics
```

# dataset 1


Loading dataset


```
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

preparing data

```
te1=TransactionEncoder()
te_ary1=te1.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df1=pd.DataFrame(te_ary1, columns=te1.columns_)  #Creating a new DataFrame from our Numpy array
df1
```

```
df1.info()
```

```
df1_transp=df1.transpose().reset_index()
```

```
df1_transp
```

apriori algorithm

```
freq_items1=apriori(df1, min_support=0.6, use_colnames=True)
freq_items1
```

```
association_rules(freq_items1,metric="confidence",min_threshold=0.7) 
```

```
association_rules(freq_items1,metric="lift",min_threshold=1.25) 
```

# Dataset 2

Loading dataset

```
df2=pd.read_csv('Market_Basket_Optimisation.csv')
```

Data exploration

```
df2.info()
```

data preparation

```
liste = []  
for i in range(7500):  
    liste.append([str(df2.values[i,j]) for j in range(20)])
```

```
te=TransactionEncoder()
te_ary=te.fit(liste).transform(liste)
df2=pd.DataFrame(te_ary, columns=te.columns_)
df2.head()
```

```
df2=df2.drop('nan', axis=1)
```

priori algorithm

```
freq_items2_1=apriori(df2, min_support=0.01, use_colnames=True)
freq_items2_1
```

```
freq_items2_2=apriori(df2, min_support=0.1, use_colnames=True)
freq_items2_2
```

```
association_rules(freq_items2_1,metric="confidence",min_threshold=0.5) 
```

```
association_rules(freq_items2_1,metric="lift",min_threshold=1.25) 
```
